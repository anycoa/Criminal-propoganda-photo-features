```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('Cleaned_photo_data.xlsx')
keywords_data = []
value_col = 'Unnamed: 0'

columns = list(df.columns)
phid_list = []
date_list = []
title_list = []
keywords_list = []
keywords_data = []

title_index_list = df.index[df[value_col] == 'View'].tolist()
for value in title_index_list:
    title_list.append(df[value_col][value+1])

phid_index_list = df.index[df[value_col] == 'Photo ID #'].tolist()


for value in phid_index_list:
    phid_list.append(df[value_col][value+1])
    date_list.append(df[value_col][value+2])

keywords_index_list = df.index[df[value_col] == 'Contains'].tolist()

idx_start = list(df.index[df[value_col] == 'Contains'])
idx_end = list(df.index[df[value_col] == 'View'])
df[value_col] = df[value_col].fillna('NaN')

value = 0

for value in range(0,len(idx_start)):
    s_value = idx_start[value] +1
    if s_value <= idx_start[-1]:
        e_value = idx_end[value+1]
        temp = list(df[value_col][s_value:e_value])
    else:
        s_value = idx_start[len(idx_start)-1] +1
        temp = list(df[value_col][s_value:])
    
    if 'NaN' in temp:
        indexes = [i for i, s in enumerate(temp) if 'NaN' in s][0]
        temp = temp[:indexes]

    if 'Run Search' in temp:
        temp.remove("Run Search")

    if 'Sort Results' in temp:
        temp.remove("Sort Results")

    keywords_data.append(",".join(temp))


df_rs = pd.DataFrame.from_dict({'Photo ID #':phid_list, 'Date': date_list,'Title':title_list,'Keywords': keywords_data},orient='index')
df_rs = df_rs.transpose()
```
